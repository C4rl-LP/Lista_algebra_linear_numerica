```python
from matplotlib import pyplot as plt
```

Definindo os dois polinômios

```python

def polinomio_com_coeficientes(x, coef = [-512, 2304, -4608, 5376,-4032, 2016, -672, 144,-18, 1] ):
    return [sum(coef[i] * xj**i for i in range(len(coef))) for xj in x]

def polinomio_compacto(x,b=2,n=9):
    return [(xi-b)**n for xi in x]
```

Gerando o espaço onde sos valores do polinômio serão avaliados


```python
espaço = [1.920 + i*.001 for i in range(1+int(2080-1920))] #fazemos 2080- 1920 ao inves de (2.080 - 1.920)/.001
print(espaço[-1])
```

```python
plt.plot(espaço, polinomio_com_coeficientes(espaço), label = 'Com todos os coeficientes')
plt.plot(espaço, polinomio_compacto(espaço), label = '$(x-2)^9$')
plt.legend()
plt.show()
```

```python
plt.plot(espaço, [polinomio_com_coeficientes(espaço)[i] -polinomio_compacto(espaço)[i] for i in range(len(espaço))] , label = '$P_1(x) - P_2(x)$')

plt.legend()
plt.show()
```
